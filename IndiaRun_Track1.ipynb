```python
# ============================================================
# EvidenceRank AI v5 - Intelligent Candidate Discovery & Ranking
# Track 1: Redrob Data & AI Challenge
# Goal: Rank Top 100 candidates for Senior AI Engineer - Founding Team
# Output: candidate_id, rank, score, reasoning
# ============================================================

import json
import re
import math
import heapq
import csv
from datetime import datetime
from collections import Counter

# ------------------------------------------------------------
# 0. FILE CONFIGURATION
# ------------------------------------------------------------

# Colab saved your uploaded file with this name.
CANDIDATES_FILE = "candidates (1).jsonl"

# Final CSV output file.
OUTPUT_FILE = "EvidenceRankAI_top100_submission.csv"

# Current date assumed from challenge period.
# Used for recency scoring.
TODAY = datetime(2026, 6, 12)

# Number of candidates retained after initial scoring for stricter re-ranking.
TOP_K_FOR_RERANK = 500

# Final required submission size.
FINAL_TOP_N = 100


# ============================================================
# 1. JD INTENT PROFILE
# ============================================================
# This converts the Senior AI Engineer JD into structured hiring intent.
# It separates must-have, preferred, hidden-gem, and risk/disqualifier signals.
# ============================================================

JD_PROFILE = {
    "role": "Senior AI Engineer - Founding Team",

    # Titles that are most relevant for this role.
    "target_titles": {
        "ai engineer", "ml engineer", "machine learning engineer",
        "applied ml engineer", "data scientist", "nlp engineer",
        "search engineer", "recommendation engineer", "software engineer",
        "backend engineer", "full stack developer", "cloud engineer",
        "devops engineer", "mlops engineer"
    },

    # Titles that are usually poor matches for this JD.
    "negative_titles": {
        "marketing manager", "content writer", "graphic designer",
        "sales executive", "accountant", "civil engineer",
        "mechanical engineer", "hr manager", "customer support",
        "operations manager", "business analyst", "project manager"
    },

    # Critical skills explicitly aligned with the JD.
    "critical_skills": {
        "python", "information retrieval", "information retrieval systems",
        "semantic search", "vector search", "vector representations",
        "sentence transformers", "rag", "search infrastructure",
        "search backend", "learning to rank", "bm25",
        "elasticsearch", "opensearch", "pinecone", "milvus",
        "llamaindex", "haystack", "fine-tuning llms",
        "lora", "qlora", "peft", "mlops", "mlflow", "kubeflow",
        "fastapi", "pytorch", "scikit-learn"
    },

    # Strong supporting skills for production AI systems.
    "supporting_skills": {
        "aws", "gcp", "azure", "kubernetes", "docker", "kafka",
        "spark", "airflow", "data pipelines", "feature engineering",
        "microservices", "rest apis", "postgresql", "mongodb",
        "databricks", "apache flink", "apache beam", "hadoop",
        "sql", "deep learning", "nlp", "natural language processing"
    },

    # Terms that prove retrieval/ranking/search experience in career text.
    "retrieval_ranking_terms": {
        "retrieval", "ranking", "ranker", "recommendation", "recommender",
        "matching", "semantic search", "vector", "embedding",
        "information retrieval", "search relevance", "search quality",
        "personalization", "candidate discovery", "content matching",
        "marketplace matching", "feed ranking", "learning to rank",
        "bm25", "index", "indexing", "elasticsearch", "opensearch",
        "faiss", "milvus", "pinecone", "qdrant", "weaviate"
    },

    # Terms that indicate production / shipped systems.
    "production_terms": {
        "production", "deployed", "real users", "scale", "scalable",
        "latency", "monitoring", "pipeline", "platform", "infrastructure",
        "service", "api", "microservice", "on-call", "owned",
        "built", "shipped", "maintained", "designed", "implemented"
    },

    # Terms that indicate evaluation and ranking-quality thinking.
    "evaluation_terms": {
        "ndcg", "mrr", "map", "a/b testing", "offline benchmark",
        "online experiment", "evaluation", "metric", "quality",
        "precision", "recall", "regression", "benchmark"
    },

    # Hidden-gem terms catch people who built relevant systems
    # without using trendy keywords like RAG/Pinecone.
    "hidden_gem_terms": {
        "recommendation engine", "recommender system", "personalization",
        "search relevance", "feed ranking", "marketplace matching",
        "matching algorithm", "ranking system", "retrieval system",
        "information retrieval", "candidate discovery", "content matching",
        "search quality", "user-activity processing", "feature pipeline"
    },

    # Consulting companies mentioned as potential risk in the JD.
    "consulting_companies": {
        "tcs", "infosys", "wipro", "accenture", "cognizant",
        "capgemini", "mindtree"
    },

    # Product/startup-like companies in synthetic dataset.
    # These are treated as better product-environment signals.
    "product_like_companies": {
        "pied piper", "initech", "acme corp", "globex inc",
        "stark industries", "wayne enterprises", "umbrella corp",
        "hooli", "massive dynamic"
    },

    # Preferred Indian hiring locations from JD.
    "preferred_locations": {
        "pune", "noida", "delhi", "gurgaon", "mumbai",
        "hyderabad", "bengaluru", "bangalore", "delhi ncr"
    },

    # Ideal experience range from JD.
    "ideal_experience_min": 5,
    "ideal_experience_max": 9
}


# ============================================================
# 2. TEXT NORMALIZATION HELPERS
# ============================================================
# These functions make profile, skill, and career text easier to score.
# ============================================================

def safe_lower(x):
    """Return lowercase string safely even if value is None."""
    return str(x or "").lower()


def normalize_text(text):
    """Clean text for keyword matching."""
    text = safe_lower(text)
    text = re.sub(r"[^a-z0-9+#.\-/ ]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text


def contains_any(text, terms):
    """Check whether any term appears inside a text."""
    text = normalize_text(text)
    return any(term in text for term in terms)


def count_term_hits(text, terms):
    """Count number of unique terms found in text."""
    text = normalize_text(text)
    return sum(1 for term in terms if term in text)


def clamp(value, low=0.0, high=100.0):
    """Keep score within a fixed range."""
    return max(low, min(high, value))


def safe_float(x, default=0.0):
    """Convert value to float safely."""
    try:
        if x is None:
            return default
        return float(x)
    except:
        return default


def parse_date(date_str):
    """Parse date safely."""
    try:
        return datetime.strptime(str(date_str), "%Y-%m-%d")
    except:
        return None


# ============================================================
# 3. CANDIDATE TEXT AND EVIDENCE BUILDERS
# ============================================================
# EvidenceRank AI v5 uses evidence hierarchy:
# Career History > Summary/Profile > Skills Section
# ============================================================

def build_candidate_text(candidate):
    """Create separated text blocks from candidate data."""
    profile = candidate.get("profile", {})
    career_history = candidate.get("career_history", [])
    education = candidate.get("education", [])
    skills = candidate.get("skills", [])
    certs = candidate.get("certifications", [])

    # Profile text includes headline and summary.
    profile_text = " ".join([
        safe_lower(profile.get("headline")),
        safe_lower(profile.get("summary")),
        safe_lower(profile.get("current_title")),
        safe_lower(profile.get("current_industry")),
        safe_lower(profile.get("current_company"))
    ])

    # Career text is the strongest evidence source.
    career_text_parts = []
    for job in career_history:
        career_text_parts.extend([
            safe_lower(job.get("title")),
            safe_lower(job.get("company")),
            safe_lower(job.get("industry")),
            safe_lower(job.get("description"))
        ])
    career_text = " ".join(career_text_parts)

    # Skills are useful but lower-trust than career proof.
    skill_names = [safe_lower(s.get("name")) for s in skills]
    skill_text = " ".join(skill_names)

    # Education/certification context.
    education_text = " ".join([
        " ".join([
            safe_lower(e.get("institution")),
            safe_lower(e.get("degree")),
            safe_lower(e.get("field_of_study")),
            safe_lower(e.get("tier"))
        ])
        for e in education
    ])

    certification_text = " ".join([
        " ".join([
            safe_lower(c.get("name")),
            safe_lower(c.get("issuer"))
        ])
        for c in certs
    ])

    # Full text for broad matching.
    full_text = " ".join([
        profile_text,
        career_text,
        skill_text,
        education_text,
        certification_text
    ])

    return {
        "profile_text": normalize_text(profile_text),
        "career_text": normalize_text(career_text),
        "skill_text": normalize_text(skill_text),
        "education_text": normalize_text(education_text),
        "certification_text": normalize_text(certification_text),
        "full_text": normalize_text(full_text),
        "skill_names": set(skill_names)
    }


def get_current_title(candidate):
    """Get normalized current title."""
    return normalize_text(candidate.get("profile", {}).get("current_title", ""))


def get_current_company(candidate):
    """Get normalized current company."""
    return normalize_text(candidate.get("profile", {}).get("current_company", ""))


def get_location(candidate):
    """Get normalized location and country."""
    profile = candidate.get("profile", {})
    return normalize_text(profile.get("location", "") + " " + profile.get("country", ""))


# ============================================================
# 4. BROAD RECALL LAYER
# ============================================================
# This layer ensures we do not miss relevant candidates.
# It rewards broad matches across title, skills, summary, and career text.
# ============================================================

def broad_recall_score(candidate, text_blocks):
    """Score how likely candidate should enter the relevant pool."""
    title = get_current_title(candidate)
    full_text = text_blocks["full_text"]
    skill_names = text_blocks["skill_names"]

    score = 0.0

    # Title match is a strong recall signal.
    if title in JD_PROFILE["target_titles"]:
        score += 30

    # Partial title match catches titles like "Senior ML Engineer".
    if any(t in title for t in JD_PROFILE["target_titles"]):
        score += 20

    # Negative titles reduce recall unless strong evidence exists elsewhere.
    if title in JD_PROFILE["negative_titles"]:
        score -= 25

    # Critical skills overlap.
    critical_overlap = len(skill_names & JD_PROFILE["critical_skills"])
    score += min(critical_overlap * 5, 35)

    # Supporting skills overlap.
    supporting_overlap = len(skill_names & JD_PROFILE["supporting_skills"])
    score += min(supporting_overlap * 2, 20)

    # Career text retrieval/ranking evidence.
    retrieval_hits = count_term_hits(text_blocks["career_text"], JD_PROFILE["retrieval_ranking_terms"])
    score += min(retrieval_hits * 6, 30)

    # Hidden-gem terms in career or summary.
    hidden_hits = count_term_hits(full_text, JD_PROFILE["hidden_gem_terms"])
    score += min(hidden_hits * 4, 20)

    return clamp(score, 0, 100)


# ============================================================
# 5. EVIDENCE EXTRACTION & CLASSIFICATION
# ============================================================
# This is the evidence hierarchy:
# - Career history proof gets highest weight.
# - Profile/summary proof gets medium weight.
# - Skills section claim gets lowest weight.
# ============================================================

def evidence_score(candidate, text_blocks):
    """Score verified evidence strength using hierarchy."""
    career_text = text_blocks["career_text"]
    profile_text = text_blocks["profile_text"]
    skill_names = text_blocks["skill_names"]

    score = 0.0
    evidence_notes = []

    # Career history proof for retrieval/ranking/search.
    career_retrieval_hits = count_term_hits(career_text, JD_PROFILE["retrieval_ranking_terms"])
    if career_retrieval_hits > 0:
        score += min(career_retrieval_hits * 10, 35)
        evidence_notes.append("career evidence for retrieval/search/ranking")

    # Career history proof for production systems.
    career_production_hits = count_term_hits(career_text, JD_PROFILE["production_terms"])
    if career_production_hits > 0:
        score += min(career_production_hits * 4, 20)
        evidence_notes.append("production/system-building evidence")

    # Career history proof for evaluation frameworks.
    career_eval_hits = count_term_hits(career_text, JD_PROFILE["evaluation_terms"])
    if career_eval_hits > 0:
        score += min(career_eval_hits * 8, 20)
        evidence_notes.append("evaluation/ranking-quality evidence")

    # Profile/summary proof is useful but weaker than career proof.
    profile_retrieval_hits = count_term_hits(profile_text, JD_PROFILE["retrieval_ranking_terms"])
    score += min(profile_retrieval_hits * 4, 12)

    # Skills section claim is included but lower weight.
    critical_skill_overlap = len(skill_names & JD_PROFILE["critical_skills"])
    supporting_skill_overlap = len(skill_names & JD_PROFILE["supporting_skills"])
    score += min(critical_skill_overlap * 3, 18)
    score += min(supporting_skill_overlap * 1.5, 10)

    return clamp(score, 0, 100), evidence_notes


# ============================================================
# 6. MUST-HAVE GATE & CORE FIT CHECK
# ============================================================
# Candidate must satisfy 2-3 major requirement categories.
# This avoids pure keyword-stuffing candidates.
# ============================================================

def must_have_gate(candidate, text_blocks):
    """Evaluate whether candidate satisfies enough core requirements."""
    skill_names = text_blocks["skill_names"]
    career_text = text_blocks["career_text"]
    full_text = text_blocks["full_text"]

    satisfied = []

    # Requirement 1: Retrieval/search/ranking.
    if count_term_hits(full_text, JD_PROFILE["retrieval_ranking_terms"]) >= 2:
        satisfied.append("retrieval/search/ranking")

    # Requirement 2: Embeddings/vector DB/search infra.
    vector_terms = {
        "vector search", "vector representations", "sentence transformers",
        "milvus", "pinecone", "opensearch", "elasticsearch", "semantic search",
        "embeddings", "embedding"
    }
    if len(skill_names & vector_terms) > 0 or count_term_hits(full_text, vector_terms) > 0:
        satisfied.append("embedding/vector/search infrastructure")

    # Requirement 3: Python/production engineering.
    if "python" in skill_names or "python" in full_text:
        satisfied.append("python")

    # Requirement 4: Production deployment.
    if count_term_hits(career_text, JD_PROFILE["production_terms"]) >= 3:
        satisfied.append("production/system deployment")

    # Requirement 5: Evaluation/ranking metrics.
    if count_term_hits(full_text, JD_PROFILE["evaluation_terms"]) > 0:
        satisfied.append("evaluation metrics")

    # Requirement 6: Relevant experience range.
    yoe = safe_float(candidate.get("profile", {}).get("years_of_experience", 0))
    if 4 <= yoe <= 12:
        satisfied.append("experience range")

    # Core pass if candidate satisfies at least 2 requirements.
    passed = len(satisfied) >= 2

    return passed, satisfied


# ============================================================
# 7. TECHNICAL FIT SCORER
# ============================================================
# This is the main JD-fit score.
# It rewards AI/ML, retrieval, ranking, search, production, and Python.
# ============================================================

def technical_fit_score(candidate, text_blocks):
    """Compute technical fit score from skills, title, career evidence, and experience."""
    profile = candidate.get("profile", {})
    skill_names = text_blocks["skill_names"]
    title = get_current_title(candidate)
    career_text = text_blocks["career_text"]
    full_text = text_blocks["full_text"]

    score = 0.0
    strengths = []

    # Title relevance.
    if title in JD_PROFILE["target_titles"]:
        score += 18
        strengths.append(f"relevant current title: {profile.get('current_title')}")
    elif any(t in title for t in JD_PROFILE["target_titles"]):
        score += 12
        strengths.append(f"partially relevant title: {profile.get('current_title')}")
    elif title in JD_PROFILE["negative_titles"]:
        score -= 15

    # Experience range.
    yoe = safe_float(profile.get("years_of_experience", 0))
    if JD_PROFILE["ideal_experience_min"] <= yoe <= JD_PROFILE["ideal_experience_max"]:
        score += 15
        strengths.append(f"{yoe:.1f} years experience within target range")
    elif 4 <= yoe < JD_PROFILE["ideal_experience_min"]:
        score += 9
        strengths.append(f"{yoe:.1f} years experience near target range")
    elif 9 < yoe <= 12:
        score += 8
        strengths.append(f"{yoe:.1f} years experience with senior depth")
    else:
        score += 2

    # Critical skill overlap.
    critical_overlap = skill_names & JD_PROFILE["critical_skills"]
    score += min(len(critical_overlap) * 4, 28)
    if critical_overlap:
        strengths.append("critical skills: " + ", ".join(sorted(list(critical_overlap))[:5]))

    # Supporting skill overlap.
    supporting_overlap = skill_names & JD_PROFILE["supporting_skills"]
    score += min(len(supporting_overlap) * 2, 16)

    # Career proof for retrieval/ranking.
    retrieval_hits = count_term_hits(career_text, JD_PROFILE["retrieval_ranking_terms"])
    score += min(retrieval_hits * 8, 24)
    if retrieval_hits:
        strengths.append("career evidence in search/retrieval/ranking")

    # Production system proof.
    production_hits = count_term_hits(career_text, JD_PROFILE["production_terms"])
    score += min(production_hits * 3, 15)
    if production_hits >= 3:
        strengths.append("production/system-building experience")

    # Evaluation proof.
    eval_hits = count_term_hits(full_text, JD_PROFILE["evaluation_terms"])
    score += min(eval_hits * 5, 10)
    if eval_hits:
        strengths.append("evaluation or ranking-quality exposure")

    return clamp(score, 0, 100), strengths


# ============================================================
# 8. BEHAVIORAL HIREABILITY SCORER
# ============================================================
# Uses Redrob signals to answer:
# "Can this person realistically be hired/contacted?"
# ============================================================

def behavioral_hireability_score(candidate):
    """Compute hireability from Redrob behavioral signals."""
    s = candidate.get("redrob_signals", {})
    score = 0.0
    notes = []

    # Open-to-work is a direct availability signal.
    if s.get("open_to_work_flag", False):
        score += 15
        notes.append("open to work")

    # Last active recency.
    last_active = parse_date(s.get("last_active_date"))
    if last_active:
        days_since_active = (TODAY - last_active).days
        if days_since_active <= 30:
            score += 18
            notes.append("recently active")
        elif days_since_active <= 90:
            score += 12
        elif days_since_active <= 180:
            score += 6
        else:
            score -= 8

    # Recruiter response rate.
    response_rate = safe_float(s.get("recruiter_response_rate", 0))
    score += response_rate * 18
    if response_rate >= 0.65:
        notes.append("strong recruiter response rate")

    # Response time.
    response_time = safe_float(s.get("avg_response_time_hours", 999))
    if response_time <= 24:
        score += 10
        notes.append("fast response time")
    elif response_time <= 72:
        score += 6
    elif response_time <= 168:
        score += 2
    else:
        score -= 5

    # Interview completion rate.
    interview_rate = safe_float(s.get("interview_completion_rate", 0))
    score += interview_rate * 12

    # Offer acceptance rate.
    offer_rate = safe_float(s.get("offer_acceptance_rate", -1))
    if offer_rate >= 0:
        score += offer_rate * 8

    # Notice period.
    notice = safe_float(s.get("notice_period_days", 180))
    if notice <= 30:
        score += 10
        notes.append("short notice period")
    elif notice <= 60:
        score += 5
    elif notice <= 90:
        score += 1
    else:
        score -= 8

    # Recruiter engagement signals.
    saved = safe_float(s.get("saved_by_recruiters_30d", 0))
    views = safe_float(s.get("profile_views_received_30d", 0))
    searches = safe_float(s.get("search_appearance_30d", 0))

    score += min(saved * 1.5, 8)
    score += min(views / 10, 5)
    score += min(searches / 100, 5)

    # Verification signals.
    if s.get("verified_email", False):
        score += 2
    if s.get("verified_phone", False):
        score += 2
    if s.get("linkedin_connected", False):
        score += 2

    return clamp(score, 0, 100), notes


# ============================================================
# 9. RISK & TRAP DETECTION ENGINE
# ============================================================
# Penalizes keyword stuffing, suspicious title-skill mismatches,
# consulting-only profiles, inactive candidates, and weak evidence.
# ============================================================

def risk_score(candidate, text_blocks, technical_score, evidence_score_value):
    """Compute risk score. Higher means worse."""
    profile = candidate.get("profile", {})
    career_history = candidate.get("career_history", [])
    skills = candidate.get("skills", [])
    signals = candidate.get("redrob_signals", {})

    title = get_current_title(candidate)
    current_company = get_current_company(candidate)
    skill_names = text_blocks["skill_names"]
    career_text = text_blocks["career_text"]
    full_text = text_blocks["full_text"]

    risk = 0.0
    concerns = []

    # Risk 1: negative title with many AI keywords.
    ai_skill_count = len(skill_names & (JD_PROFILE["critical_skills"] | JD_PROFILE["supporting_skills"]))
    if title in JD_PROFILE["negative_titles"] and ai_skill_count >= 4:
        risk += 25
        concerns.append("title-skill mismatch")

    # Risk 2: keyword stuffing.
    # Many advanced AI skills but little/no career proof.
    if ai_skill_count >= 8 and evidence_score_value < 25:
        risk += 22
        concerns.append("many AI skills with weak career evidence")

    # Risk 3: consulting-only career.
    companies = {normalize_text(job.get("company", "")) for job in career_history}
    if companies and companies.issubset(JD_PROFILE["consulting_companies"]):
        risk += 12
        concerns.append("consulting-only background")

    # Risk 4: current company is consulting and no strong product evidence.
    if current_company in JD_PROFILE["consulting_companies"] and not contains_any(career_text, JD_PROFILE["product_like_companies"]):
        risk += 6

    # Risk 5: very low profile completeness.
    completeness = safe_float(signals.get("profile_completeness_score", 0))
    if completeness < 40:
        risk += 10
        concerns.append("low profile completeness")

    # Risk 6: inactive profile.
    last_active = parse_date(signals.get("last_active_date"))
    if last_active:
        days_since_active = (TODAY - last_active).days
        if days_since_active > 180:
            risk += 15
            concerns.append("inactive profile")
        elif days_since_active > 90:
            risk += 8

    # Risk 7: low recruiter response rate.
    response_rate = safe_float(signals.get("recruiter_response_rate", 0))
    if response_rate < 0.2:
        risk += 12
        concerns.append("low recruiter response rate")

    # Risk 8: very long notice period.
    notice = safe_float(signals.get("notice_period_days", 180))
    if notice >= 120:
        risk += 12
        concerns.append("long notice period")
    elif notice >= 90:
        risk += 6

    # Risk 9: no GitHub/public engineering signal despite technical claim.
    github = safe_float(signals.get("github_activity_score", -1))
    if github < 0 and technical_score > 70:
        risk += 6

    # Risk 10: suspicious experience-skill mismatch.
    yoe = safe_float(profile.get("years_of_experience", 0))
    advanced_skills = sum(
        1 for sk in skills
        if safe_lower(sk.get("proficiency")) == "advanced"
    )
    if yoe < 2 and advanced_skills >= 6:
        risk += 20
        concerns.append("low experience with too many advanced skills")

    return clamp(risk, 0, 100), concerns


# ============================================================
# 10. HIDDEN-GEM DISCOVERY BOOSTER
# ============================================================
# Boost candidates with real system evidence even without trendy keywords.
# ============================================================

def hidden_gem_boost(candidate, text_blocks, technical_score, risk):
    """Boost candidates with strong relevant career evidence but weaker keyword visibility."""
    career_text = text_blocks["career_text"]
    skill_names = text_blocks["skill_names"]
    signals = candidate.get("redrob_signals", {})

    boost = 0.0
    notes = []

    # Strong hidden-gem career terms.
    hidden_hits = count_term_hits(career_text, JD_PROFILE["hidden_gem_terms"])
    if hidden_hits > 0:
        boost += min(hidden_hits * 4, 14)
        notes.append("hidden-gem system evidence")

    # Good GitHub + decent technical score.
    github = safe_float(signals.get("github_activity_score", -1))
    if github >= 50 and technical_score >= 45:
        boost += 5
        notes.append("strong GitHub activity")

    # Low recruiter visibility but strong technical signal.
    saved = safe_float(signals.get("saved_by_recruiters_30d", 0))
    views = safe_float(signals.get("profile_views_received_30d", 0))
    if technical_score >= 65 and saved <= 2 and views <= 20:
        boost += 5
        notes.append("under-discovered technical profile")

    # Avoid boosting high-risk profiles.
    if risk >= 50:
        boost *= 0.3
    elif risk >= 30:
        boost *= 0.6

    return clamp(boost, 0, 20), notes


# ============================================================
# 11. LOCATION & NOTICE FIT
# ============================================================
# Small score for logistical suitability.
# ============================================================

def location_notice_score(candidate):
    """Score location and notice period fit."""
    location = get_location(candidate)
    signals = candidate.get("redrob_signals", {})

    score = 0.0
    notes = []

    # Preferred location.
    if any(loc in location for loc in JD_PROFILE["preferred_locations"]):
        score += 8
        notes.append("preferred location")
    elif candidate.get("profile", {}).get("country", "").lower() == "india":
        score += 4

    # Relocation willingness.
    if signals.get("willing_to_relocate", False):
        score += 4
        notes.append("willing to relocate")

    # Work mode.
    mode = normalize_text(signals.get("preferred_work_mode", ""))
    if mode in {"hybrid", "flexible"}:
        score += 3

    # Notice period.
    notice = safe_float(signals.get("notice_period_days", 180))
    if notice <= 30:
        score += 5
    elif notice <= 60:
        score += 3
    elif notice <= 90:
        score += 1

    return clamp(score, 0, 20), notes


# ============================================================
# 12. SKILL ASSESSMENT SCORE
# ============================================================
# Redrob skill assessments are more trustworthy than claimed skills.
# ============================================================

def assessment_score(candidate):
    """Score skill assessment results for relevant AI/search skills."""
    assessments = candidate.get("redrob_signals", {}).get("skill_assessment_scores", {})
    if not assessments:
        return 0.0, []

    total = 0.0
    count = 0
    notes = []

    for skill, value in assessments.items():
        skill_norm = normalize_text(skill)
        val = safe_float(value, 0)

        # Relevant assessed skills count more.
        if skill_norm in JD_PROFILE["critical_skills"] or skill_norm in JD_PROFILE["supporting_skills"]:
            total += val
            count += 1
            if val >= 75:
                notes.append(f"strong assessment in {skill}")

    if count == 0:
        return 0.0, notes

    avg = total / count
    return clamp(avg, 0, 100), notes


# ============================================================
# 13. FINAL SCORE FORMULA
# ============================================================
# v5 formula:
# Final Score = Fit × (1 - Risk/100)
# Fit combines technical, retrieval proof, production fit,
# behavioral hireability, location/notice, assessment, hidden-gem boost.
# ============================================================

def score_candidate(candidate):
    """Full EvidenceRank AI v5 scoring pipeline for one candidate."""
    text_blocks = build_candidate_text(candidate)

    # Broad recall layer.
    recall = broad_recall_score(candidate, text_blocks)

    # Evidence extraction layer.
    ev_score, ev_notes = evidence_score(candidate, text_blocks)

    # Must-have gate.
    gate_passed, gate_signals = must_have_gate(candidate, text_blocks)

    # Technical fit.
    tech_score, tech_notes = technical_fit_score(candidate, text_blocks)

    # Behavioral hireability.
    hireability, hire_notes = behavioral_hireability_score(candidate)

    # Skill assessment score.
    assess_score, assess_notes = assessment_score(candidate)

    # Risk score.
    risk, risk_notes = risk_score(candidate, text_blocks, tech_score, ev_score)

    # Hidden-gem boost.
    hidden_boost, hidden_notes = hidden_gem_boost(candidate, text_blocks, tech_score, risk)

    # Location and notice fit.
    loc_notice, loc_notes = location_notice_score(candidate)

    # Product/production fit derived from evidence and production terms.
    production_fit = min(count_term_hits(text_blocks["career_text"], JD_PROFILE["production_terms"]) * 5, 100)

    # Retrieval/ranking proof derived from career evidence.
    retrieval_proof = min(count_term_hits(text_blocks["career_text"], JD_PROFILE["retrieval_ranking_terms"]) * 10, 100)

    # If must-have gate fails, apply strong but not total penalty.
    # This allows hidden gems but prevents weak profiles from dominating.
    gate_multiplier = 1.0 if gate_passed else 0.65

    # Weighted fit score.
    fit_score = (
        tech_score * 0.35 +
        ev_score * 0.18 +
        retrieval_proof * 0.14 +
        production_fit * 0.08 +
        hireability * 0.10 +
        assess_score * 0.07 +
        loc_notice * 0.04 +
        recall * 0.04
    )

    # Hidden gem boost is additive before risk multiplier.
    fit_score += hidden_boost

    # Apply gate multiplier.
    fit_score *= gate_multiplier

    # Apply risk multiplier.
    final_score = fit_score * (1 - risk / 100)

    # Keep final score inside clean range.
    final_score = clamp(final_score, 0, 100)

    # Store explanation data for reasoning generation.
    details = {
        "candidate_id": candidate.get("candidate_id"),
        "final_score": final_score,
        "fit_score": fit_score,
        "recall_score": recall,
        "evidence_score": ev_score,
        "technical_score": tech_score,
        "retrieval_proof": retrieval_proof,
        "production_fit": production_fit,
        "hireability_score": hireability,
        "assessment_score": assess_score,
        "risk_score": risk,
        "hidden_gem_boost": hidden_boost,
        "location_notice_score": loc_notice,
        "gate_passed": gate_passed,
        "gate_signals": gate_signals,
        "strengths": tech_notes + ev_notes + hire_notes + assess_notes + hidden_notes + loc_notes,
        "concerns": risk_notes
    }

    return final_score, details


# ============================================================
# 14. PAIRWISE TOP-500 RE-RANKER / CALIBRATION
# ============================================================
# This re-scores the strongest candidates more strictly.
# It improves top-10/top-50 quality.
# ============================================================

def rerank_top_candidates(scored_candidates):
    """Apply stricter calibration to top candidates."""
    reranked = []

    for score, candidate, details in scored_candidates:
        adjusted = score

        # Prefer candidates who passed must-have gate.
        if details["gate_passed"]:
            adjusted += 3
        else:
            adjusted -= 6

        # Strongly reward actual retrieval/ranking proof.
        if details["retrieval_proof"] >= 30:
            adjusted += 4

        # Reward strong technical fit with low risk.
        if details["technical_score"] >= 70 and details["risk_score"] <= 25:
            adjusted += 4

        # Penalize high-risk candidates even if technical score is high.
        if details["risk_score"] >= 45:
            adjusted -= 8
        elif details["risk_score"] >= 30:
            adjusted -= 4

        # Strong hireability matters because JD mentions realistic hiring.
        if details["hireability_score"] >= 60:
            adjusted += 3

        # Boost hidden gems only if risk is acceptable.
        if details["hidden_gem_boost"] >= 8 and details["risk_score"] <= 30:
            adjusted += 2

        adjusted = clamp(adjusted, 0, 100)

        details["reranked_score"] = adjusted
        reranked.append((adjusted, candidate, details))

    # Sort by:
    # 1. adjusted score descending
    # 2. risk ascending
    # 3. candidate_id ascending for deterministic tie-break
    reranked.sort(
        key=lambda x: (
            -x[0],
            x[2]["risk_score"],
            x[1]["candidate_id"]
        )
    )

    return reranked


# ============================================================
# 15. REASONING GENERATOR & VALIDATOR
# ============================================================
# Reasoning must use actual candidate facts.
# It should mention:
# - relevant strength
# - JD connection
# - concern if applicable
# - no hallucinated facts
# ============================================================

def generate_reasoning(candidate, details):
    """Generate concise, factual reasoning for submission CSV."""
    profile = candidate.get("profile", {})
    signals = candidate.get("redrob_signals", {})

    title = profile.get("current_title", "candidate")
    yoe = safe_float(profile.get("years_of_experience", 0))
    location = profile.get("location", "unknown location")

    strengths = details.get("strengths", [])
    concerns = details.get("concerns", [])
    gate_signals = details.get("gate_signals", [])

    # Pick strongest available fact.
    if strengths:
        strength_text = strengths[0]
    elif gate_signals:
        strength_text = "matches core requirement: " + gate_signals[0]
    else:
        strength_text = "has partial alignment with the role requirements"

    # Mention behavioral signal only if clearly positive.
    behavior_bits = []
    if signals.get("open_to_work_flag", False):
        behavior_bits.append("open to work")
    if safe_float(signals.get("recruiter_response_rate", 0)) >= 0.6:
        behavior_bits.append("strong recruiter response rate")
    if safe_float(signals.get("notice_period_days", 180)) <= 30:
        behavior_bits.append("short notice period")

    if behavior_bits:
        behavior_text = " Also shows hireability signals: " + ", ".join(behavior_bits[:2]) + "."
    else:
        behavior_text = ""

    # Mention one concern if present.
    if concerns:
        concern_text = f" Concern: {concerns[0]}."
    elif details["risk_score"] > 25:
        concern_text = " Concern: moderate profile risk after validation."
    else:
        concern_text = ""

    # Final reasoning, kept concise.
    reasoning = (
        f"{title} with {yoe:.1f} years of experience in {location}; "
        f"{strength_text} for the Senior AI Engineer role."
        f"{behavior_text}{concern_text}"
    )

    # Clean commas/newlines to keep CSV neat.
    reasoning = reasoning.replace("\n", " ").replace("\r", " ")
    reasoning = re.sub(r"\s+", " ", reasoning).strip()

    # Keep reasoning compact.
    return reasoning[:450]


# ============================================================
# 16. RUN FULL PIPELINE ON 100,000 CANDIDATES
# ============================================================
# Uses a heap to avoid storing all 100K full candidate objects.
# Keeps only top TOP_K_FOR_RERANK candidates by initial score.
# ============================================================

top_heap = []
processed = 0

print("Starting EvidenceRank AI v5 ranking pipeline...")
print(f"Reading file: {CANDIDATES_FILE}")

with open(CANDIDATES_FILE, "r", encoding="utf-8") as f:
    for line in f:
        processed += 1

        # Parse one candidate JSON record.
        candidate = json.loads(line)

        # Score candidate using full v5 pipeline.
        score, details = score_candidate(candidate)

        # Store tuple.
        # Heap keeps smallest score at root, so we can maintain top K efficiently.
        item = (score, candidate["candidate_id"], candidate, details)

        if len(top_heap) < TOP_K_FOR_RERANK:
            heapq.heappush(top_heap, item)
        else:
            # Replace weakest candidate if current candidate is stronger.
            if score > top_heap[0][0]:
                heapq.heapreplace(top_heap, item)

        # Progress log every 10K candidates.
        if processed % 10000 == 0:
            print(f"Processed {processed:,} candidates... current heap size = {len(top_heap)}")

print(f"Completed initial scoring for {processed:,} candidates.")


# ============================================================
# 17. TOP-500 RE-RANKING
# ============================================================

# Convert heap to sorted list.
top_candidates_initial = [
    (score, candidate, details)
    for score, cid, candidate, details in top_heap
]

print("Applying Top-500 pairwise-style calibration and re-ranking...")

reranked_candidates = rerank_top_candidates(top_candidates_initial)

# Keep final top 100.
final_candidates = reranked_candidates[:FINAL_TOP_N]


# ============================================================
# 18. EXPORT FINAL SUBMISSION CSV
# ============================================================

print(f"Writing final submission to: {OUTPUT_FILE}")

with open(OUTPUT_FILE, "w", encoding="utf-8", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Required header exactly.
    writer.writerow(["candidate_id", "rank", "score", "reasoning"])

    for rank, (score, candidate, details) in enumerate(final_candidates, start=1):
        candidate_id = candidate["candidate_id"]

        # Submission score should be monotonically non-increasing.
        # Use reranked score and round for readability.
        final_score = round(float(score), 6)

        # Generate factual reasoning.
        reasoning = generate_reasoning(candidate, details)

        writer.writerow([candidate_id, rank, final_score, reasoning])


# ============================================================
# 19. BASIC VALIDATION CHECKS
# ============================================================
# These checks mirror the official submission requirements.
# ============================================================

print("Running local validation checks...")

with open(OUTPUT_FILE, "r", encoding="utf-8") as f:
    rows = list(csv.DictReader(f))

# Check exactly 100 rows.
assert len(rows) == 100, f"Expected 100 rows, found {len(rows)}"

# Check ranks 1-100 exactly.
ranks = [int(r["rank"]) for r in rows]
assert ranks == list(range(1, 101)), "Ranks must be exactly 1 to 100"

# Check unique candidate IDs.
candidate_ids = [r["candidate_id"] for r in rows]
assert len(candidate_ids) == len(set(candidate_ids)), "Duplicate candidate IDs found"

# Check score descending.
scores = [float(r["score"]) for r in rows]
assert all(scores[i] >= scores[i+1] for i in range(len(scores)-1)), "Scores are not non-increasing"

# Check reasoning not empty.
assert all(r["reasoning"].strip() for r in rows), "Empty reasoning found"

print("✅ Submission format looks valid.")
print(f"✅ Final file created: {OUTPUT_FILE}")


# ============================================================
# 20. PREVIEW TOP 10
# ============================================================

print("\nTop 10 preview:\n")

for r in rows[:10]:
    print(f"Rank {r['rank']} | {r['candidate_id']} | Score: {r['score']}")
    print(f"Reasoning: {r['reasoning']}")
    print("-" * 80)


# ============================================================
# 21. DOWNLOAD OUTPUT FILE
# ============================================================
# Uncomment this in Colab if you want automatic download.
# ============================================================

from google.colab import files
files.download(OUTPUT_FILE)
```
